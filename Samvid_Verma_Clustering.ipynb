```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score, silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

customers_path = 'Customers.csv'
transactions_path = 'Transactions.csv'
products_path = 'Products.csv'

customers_df = pd.read_csv(customers_path)
transactions_df = pd.read_csv(transactions_path)
products_df = pd.read_csv(products_path)

customer_transaction_features = transactions_df.groupby("CustomerID").agg({
    "TotalValue": "sum",             
    "TransactionID": "count",        
    "ProductID": "nunique"           
}).rename(columns={
    "TotalValue": "TotalSpend",
    "TransactionID": "NumTransactions",
    "ProductID": "UniqueProducts"
}).reset_index()

customer_profiles = customers_df.merge(customer_transaction_features, on="CustomerID", how="inner")

features = customer_profiles[["TotalSpend", "NumTransactions", "UniqueProducts"]]

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

cluster_results = {}
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(features_scaled)
    db_index = davies_bouldin_score(features_scaled, labels)
    silhouette_avg = silhouette_score(features_scaled, labels)
    cluster_results[n_clusters] = {"DBIndex": db_index, "SilhouetteScore": silhouette_avg, "Labels": labels}

best_n_clusters = min(cluster_results, key=lambda k: cluster_results[k]["DBIndex"])
best_labels = cluster_results[best_n_clusters]["Labels"]
best_db_index = cluster_results[best_n_clusters]["DBIndex"]
best_silhouette = cluster_results[best_n_clusters]["SilhouetteScore"]

customer_profiles["Cluster"] = best_labels

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=features_scaled[:, 0], y=features_scaled[:, 1],
    hue=best_labels, palette="viridis", legend="full"
)
plt.title(f"Customer Segmentation (Best Clusters: {best_n_clusters})")
plt.xlabel("Feature 1 (Scaled)")
plt.ylabel("Feature 2 (Scaled)")
plt.legend(title="Clusters")
plt.tight_layout()

plot_path = "Cluster_Visualization.png"
plt.savefig(plot_path)
plt.show()

clustering_report_path = "Clustering_Report.csv"
customer_profiles.to_csv(clustering_report_path, index=False)

print("Best Number of Clusters:", best_n_clusters)
print("Best DB Index:", best_db_index)
print("Best Silhouette Score:", best_silhouette)
print("Clustering report saved to:", clustering_report_path)
print("Cluster visualization saved to:", plot_path)
```
